```python
from scipy.special import jv
import matplotlib.pyplot as plt
import numpy as np
k=np.linspace(0,10,1000)
J=jv(1/2,k)
j=jv(1/2,k)*np.sqrt(np.pi/2/k)
plt.plot(k,J)
plt.plot(k,j)
```
